```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load training data
train_data = pd.read_csv(r'C:\DSClean\NLP_Disaster_Tweets\data\interim\train.csv')
```

```python
# Explore and show distribution of target variable in bar chart showing value count in bar
target = train_data['target'].value_counts()
target.plot(kind='bar', color=['blue', 'red'], title='Distribution of target variable', ylabel='Count', xlabel='Target')
```

```python
# Get total, average, max, min number of words and vocabulary size in the training data and print as table
train_data['text'] = train_data['text'].astype(str)
train_data['word_count'] = train_data['text'].apply(lambda x: len(x.split()))
total_words = train_data['word_count'].sum()
average_words = train_data['word_count'].mean()
max_words = train_data['word_count'].max()
min_words = train_data['word_count'].min()
vocabulary_size = len(set(' '.join(train_data['text']).split()))
word_stats = pd.DataFrame({'Total words': [total_words], 'Average words': [average_words], 'Max words': [max_words],
                           'Min words': [min_words], 'Vocabulary size': [vocabulary_size]})
print(word_stats)
```
